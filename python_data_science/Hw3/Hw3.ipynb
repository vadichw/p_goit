```python
def hypothesis(X, theta):
    h = np.dot(X, theta)
    return h

# Приклад використання функції гіпотези
X = np.array([[1, 2], [1, 3], [1, 4]])
theta = np.array([[3], [2]])
predictions = hypothesis(X, theta)
print(predictions)
```

```python
def loss_cost(X, y, theta):
    # Кількість спостережень
    m = len(y)

    # Обчислюємо прогнози за допомогою функції гіпотези
    h = np.dot(X, theta)

    # Обчислюємо втрати (середньоквадратична помилка)
    J = (1 / (2 * m)) * np.sum(np.square(h - y))

    return J


X = np.array([[1, 2], [1, 3], [1, 4]])  # Припустимо, що у нас є три спостереження та дві ознаки, і додаємо стовпець з одиницями
y = np.array([[4], [5], [6]])          # Відповіді для кожного спостереження
theta = np.array([[1], [2]])           # Параметри моделі (включаючи параметр зсуву)
cost = loss_cost(X, y, theta)
print(cost)
```

```python
def gradient_descent_step(X, y, theta, learning_rate):

    # Кількість спостережень
    m = len(y)

    # Обчислюємо прогнози за допомогою поточних параметрів
    h = np.dot(X, theta)

    # Обчислюємо градієнт функції втрат
    gradient = (1 / m) * np.dot(X.T, (h - y))

    # Оновлюємо параметри за допомогою градієнтного спуску
    new_theta = theta - learning_rate * gradient

    return new_theta


X = np.array([[1, 2], [1, 3], [1, 4]])  # Припустимо, що у нас є три спостереження та дві ознаки, і додаємо стовпець з одиницями
y = np.array([[4], [5], [6]])          # Відповіді для кожного спостереження
theta = np.array([[0], [0]])           # Початкові параметри моделі (включаючи параметр зсуву)
learning_rate = 0.01                   # Швидкість навчання
new_theta = gradient_descent_step(X, y, theta, learning_rate)
print(new_theta)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd

df = pd.read_csv('/content/Housing.csv')
df.head()
```

```python



# Вибираємо відповідні стовпці
prices = df['price'].values
areas = df['area'].values
bathrooms = df['bathrooms'].values
bedrooms = df['bedrooms'].values

# Формування матриці ознак (кожен рядок - це будинок, стовпці - площа, ванні, спальні)
X = np.column_stack((areas, bathrooms, bedrooms))

# Навчання моделі лінійної регресії
model = LinearRegression().fit(X, prices)

# Виведення знайдених коефіцієнтів моделі (w0, w1, w2, w3)
print("Коефіцієнти моделі:", model.intercept_, model.coef_)
```

```python
import matplotlib.pyplot as plt

# Прогноз цін на будинки
predicted_prices = model.predict(X)

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.scatter(prices, predicted_prices, color='blue', label='Прогнозовані ціни')
plt.plot([prices.min(), prices.max()], [prices.min(), prices.max()], color='red', linestyle='--', label='Ідеальна лінія')
plt.xlabel('Реальні ціни на будинки')
plt.ylabel('Прогнозовані ціни на будинки')
plt.title('Порівняння реальних та прогнозованих цін на будинки')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Обчислення середніх значень
mean_prices = np.mean(prices)
mean_areas = np.mean(areas)
mean_bathrooms = np.mean(bathrooms)
mean_bedrooms = np.mean(bedrooms)

# Обчислення коефіцієнтів кожного незалежного параметра
numerator1 = np.sum((areas - mean_areas) * (prices - mean_prices))
denominator1 = np.sum((areas - mean_areas) ** 2)
w1 = numerator1 / denominator1

numerator2 = np.sum((bathrooms - mean_bathrooms) * (prices - mean_prices))
denominator2 = np.sum((bathrooms - mean_bathrooms) ** 2)
w2 = numerator2 / denominator2

numerator3 = np.sum((bedrooms - mean_bedrooms) * (prices - mean_prices))
denominator3 = np.sum((bedrooms - mean_bedrooms) ** 2)
w3 = numerator3 / denominator3

# Обчислення коефіцієнта зсуву (intercept)
w0 = mean_prices - (w1 * mean_areas) - (w2 * mean_bathrooms) - (w3 * mean_bedrooms)

print("Коефіцієнти моделі:", w0, w1, w2, w3)
```
